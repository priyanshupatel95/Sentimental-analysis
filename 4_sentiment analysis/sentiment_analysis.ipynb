```python
import pandas as pd
data = pd.read_csv('nearly_final.csv')
data=  data.drop(['Unnamed: 0'],axis = 1)
data.head()
```

```python

my_file = open("positive-words.txt", "r")
data = my_file.read()
positive_list = data.split("\n")
my_file.close()

```

```python
import chardet
with open("negative-words.txt", 'rb') as rawdata:
    result = chardet.detect(rawdata.read(100000))
result
```

```python
my_file = open("negative-words.txt", "r",encoding="ISO-8859-1")
data = my_file.read()
negative_list = data.split("\n")
my_file.close()
```

```python
def positive_count(word):
  p_count=0
  # pronouns = ["I","we","my","ours","us"]
  for i in word.split():
    if i in positive_list:
      p_count+=1
  return p_count

data['positive_score'] = data['clean_content'].apply(lambda x :positive_count(x))
```

```python
def negative_count(word):
  n_count=0
  # pronouns = ["I","we","my","ours","us"]
  for i in word.split():
    if i in negative_list:
      n_count+=1
  return n_count

data['negative_score']=data['clean_content'].apply(lambda x :negative_count(x))
data.head()
```

```python
data.to_csv('p_n_final.csv')
```

```python
new_dataframe= pd.read_csv('p_n_final.csv')

new_dataframe.head()
```

```python
x= new_dataframe['positive_score']
y = new_dataframe['negative_score']
total_words = new_dataframe['no_of_total_words']
new_dataframe['Polarity Score'] = ((x-y)/((x+y)+0.000001))
new_dataframe['Subjectivity score'] = ((x+y)/(total_words+0.00001))
new_dataframe.head()
```

```python
xy=  pd.read_excel('final_extracted.xlsx')
new_dataframe['URL']= xy['url']
#new_dataframe = new_dataframe.drop(['POS tagged'], axis= 1)
new_dataframe.head()
```

```python
#@markdown Rearranging the final output columns to required
outdata = pd.DataFrame(columns = [])
outdata['URL_ID'] = new_dataframe['Url_id']
outdata['URL'] = new_dataframe['URL']
outdata['POSITIVE SCORE'] = new_dataframe['positive_score']
outdata['NEGATIVE SCORE'] = new_dataframe['negative_score']
outdata['POLARITY SCORE'] = new_dataframe['Polarity Score']
outdata['SUBJECTIVITY SCORE'] = new_dataframe['Subjectivity score']
outdata['AVG SENTENCE LENGTH'] = new_dataframe['Average number of Words Per Sentence']
outdata['PERCENTAGE OF COMPLEX WORDS'] = new_dataframe['complex words percentage']
outdata['FOG INDEX'] = new_dataframe['fog index']
outdata['AVG NUMBER OF WORDS PER SENTENCE'] = new_dataframe['Average number of Words Per Sentence']
outdata['COMPLEX WORD COUNT'] = new_dataframe['Complex word Count']
outdata['WORD COUNT'] = new_dataframe['no_of_total_words']
outdata['SYLLABLE PER WORD'] = new_dataframe['syllables']
outdata['PERSONAL PRONOUNS'] = new_dataframe['personal_pronouns']
outdata['AVG WORD LENGTH'] = new_dataframe['Average word length']
outdata.head()
```
