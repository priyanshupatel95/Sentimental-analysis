```python
import nltk
```

```python
import pandas as pd
data = pd.read_csv('final_no_case_change_data (1).csv')
data=  data.drop(['Unnamed: 0'],axis = 1)
data.head()
```

```python
def word_count(word):
    # word = word.lower()
    count = 0
    para_list=word.split()
    count=len(para_list)
    return count

data['no_of_total_words']=data['clean_content'].apply(lambda x :word_count(x))
data.head()
```

```python

from nltk.tokenize import word_tokenize
from nltk import pos_tag
nltk.download('stopwords')
from nltk.corpus import stopwords
nltk.download('wordnet')
from nltk.corpus import wordnet
all_stopwords = stopwords.words('english')


# POS tagger dictionary
def token_stop_pos(text):
    word = word_tokenize(text)
    newlist = []
    for word in word :
        newlist.append(tuple([word]))
    return newlist

data['POS tagged'] = data['clean_content'].apply(token_stop_pos)
data.head()
```

```python
# Average number of Words Per Sentence
col_list= ['no_of_total_words', 'no_of_sentences']
# sum specific columns
data['Average number of Words Per Sentence'] = data['no_of_total_words']//data['no_of_sentences']
data.head()
```

```python
import re
def syllable_count(word):
    # word = word.lower()
    count = 0
    para_list=word.split()
    for i in para_list:
      for ch in i:
        if re.match(r'[aeiouAEIOU]', ch):
          count+=1
      if i.endswith("es") or i.endswith("ed"):
        count-=1
    return count

data['syllables'] = data['clean_content'].apply(lambda x :syllable_count(x))
data.head()
```

```python
def syllable_count(word):
    word = word.lower()
    complex_count = 0
    para_list=word.split()
    for i in para_list:
      word_syllable_count=0
      for ch in i:
        if re.match(r'[aeiou]', ch):
          word_syllable_count+=1

      if i.endswith("es") or i.endswith("ed"):
        word_syllable_count-=1
      if(word_syllable_count>2):
        complex_count+=1
        # print(i)
    return complex_count

data['Complex word Count'] = data['clean_content'].apply(lambda x :syllable_count(x))
data.head()
```

```python

data['Average word length'] = data['no_of_characters']//data['no_of_total_words']
data.head()
```

```python
data['complex words percentage'] = (data['Complex word Count']/data['no_of_total_words'])*100
data['fog index'] = (data['complex words percentage'] +data['Average number of Words Per Sentence'])*4
data.head()
```

```python
def pronouns_count(word):
  count=0
  pronouns = ["I","we","my","ours","us"]
  for i in word.split():
    if i in pronouns:
      count+=1
  return count

data['personal_pronouns']=data['clean_content'].apply(lambda x :pronouns_count(x))
data.head()
```
